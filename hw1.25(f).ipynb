Recall in Problem 1.25 (f), we have been asked whether $x(t)=\sum_{n=-\infty}^{\infty}e^{-(2t-n)}u(2t-n)$ is a periodic signal or not.

Before answering the question, let's first use python to plot the function $x(t)$ to get a rough idea.

Let's first define a few functions:

I. define u(2t-n).

```python
def ustep(t,n):
    out=0.0
    if (2*t-n)>0.0:
        out=1.0
    return out
```

II. define e(-(2t-n)).

```python
def sig(t,n):
    return exp(-(2*t-n));
```

III. define $\sum_{n=-N}^{N}e(-(2t-n))u(2t-n)$.

```python
def summer(t,N):
    ns = range(-N,N+1)
    out=0.0
    for n in ns:
        out = out + sig(t,n)*ustep(t,n)
    return out
```

```python
ts=(np.linspace(0,10,5000))# visualize t from 0 to 10 with 5000 discrete points
```

Let's first plot $e^{-(2t-n)}u(2t-n)$ for n = 0, which means $e^{-(2t)}u(2t)$

```python
n1 = 0
x1s=[sig(t,n1)*ustep(t,n1) for t in ts]
plot(ts, x1s)
```

Now let's  plot a range of $e^{-(2t-n)}u(2t-n)$ each with a specific $n$ from -20 to 20,
and overlap one on top of each other.

```python
N = 20
for n1 in range(-N,N+1):
    x1s=[sig(t,n1)*ustep(t,n1) for t in ts]
    plot(ts, x1s)
```

Finally, let's plot $\sum_{n=-N}^{N}e(-(2t-n))u(2t-n)$ for N=20

```python
xs=[summer(t,N) for t in ts]
plot(ts, xs)
```

When N=20, it seems the function is periodic with period $0.5$ from our oberservation of $t$ from 0 to 10.

Although we only set N=20, you can change N and see similar results.

```python
import numpy as np
import scipy
```

define u(2t-n)

```python
def ustep(t,n):
    out=1.0
    if (2*t-n)<0.0:
        out=0.0
    return out
```

define e(-(2t-n))

```python
def sig(t,n):
    return exp(-(2*t-n));
```

define $\sum_{n=-\infty}^{\infty}e(-(2t-n))u(2t-n)$

```python
def summer(t,N):
    ns = range(-N,N+1)
    #print len(ns)
    out=0.0
    for n in ns:
        out = out + sig(t,n)*ustep(t,n)
    return out
```

```python
ts=(np.linspace(0,10,5000))
n=20
xs=[summer(t,n) for t in ts]
```

```python
for n1 in range(-n,n+1):
    x1s=[sig(t,n1)*ustep(t,n1) for t in ts]
    plot(ts, x1s)
```
